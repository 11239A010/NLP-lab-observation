```python
def ngram_smooth(text, n=2):
    words = text.split()
    N = len(words) - n + 1
    ngrams = [tuple(words[i:i+n]) for i in range(N)]
    V = len(set(ngrams))
    for ng in sorted(set(ngrams)):
        count = ngrams.count(ng)
        print(ng, "->", round((count+1)/(N+V), 4))

text = "I love NLP and I love Python"
ngram_smooth(text, 2)
```
