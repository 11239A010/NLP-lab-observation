```
def bigram_model(text):
    words = text.split()
    total = len(words) - 1
    seen = set()
    for i in range(total):
        pair = (words[i], words[i+1])
        if pair not in seen:
            print(pair, "->", sum(1 for j in range(total) if (words[j], words[j+1])==pair)/total)
            seen.add(pair)

text = "I love NLP and I love Python"
bigram_model(text)

```
