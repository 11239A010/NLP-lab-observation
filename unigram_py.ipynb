```python
from collections import Counter

def unigram_prob(text):
    w = text.split()
    c = Counter(w)
    N = sum(c.values())
    for k,v in c.items():
        print(k, "->", round(v/N, 4))

text = "I love NLP and I love Python"
unigram_prob(text)
```
