```python
def trigram_model(text):
    w = text.split()
    t = len(w)-2
    for i in range(t):
        print((w[i], w[i+1], w[i+2]), "->", w[i:i+3].count(w[i:i+3])/t)

text = "I love NLP and I love Python NLP"
trigram_model(text)
```
